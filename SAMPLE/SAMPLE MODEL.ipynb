```python


# ------------------------------------------------


#      SAMPLE MODEL


# ------------------------------------------------


###  INPUT ###
"""

Read CSV files from the given list of file paths and return DataFrames.

"""

import pandas as pd

input_data = pd.read_csv("2023_smartFarm_AI_hackathon_dataset.csv")
#  -------------------------------------------
###    Write codes...  ###
#      EXAMPLE CODE      #




# Read CSV files from the lists of file paths


# Now you have lists of DataFrames for each type of data 

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
from lightgbm.sklearn import LGBMRegressor






# ... (Data preprocessing code here)
input_data = input_data.drop(columns=['frmDist'])

# Split the data into training and testing sets
X = input_data[input_data.drop(columns=['outtrn_cumsum','HeatingEnergyUsage_cumsum']).columns]
Y = input_data[['outtrn_cumsum','HeatingEnergyUsage_cumsum']]

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)


# Initialize and train the LinearRegression model
# model_1 = LinearRegression()
# model_1.fit(X_train, y_train)

# model_rf = RandomForestRegressor()
# model_rf.fit(X_train, y_train)

model_lg = LGBMRegressor()
model_lg.fit(X_train, y_train[["outtrn_cumsum"]])

# model = XGBRegressor()
# model.fit(X_train, y_train)

# Predict 'y' values using the trained model_1
# y_pred_1 = model_1.predict(X_test)
# y_pred = model.predict(X_test)
# y_pred_rf = model_rf.predict(X_test)
y_pred_lg_0 = model_lg.predict(X_test)

model_lg.fit(X_train, y_train[["HeatingEnergyUsage_cumsum"]])
y_pred_lg_1 = model_lg.predict(X_test)

y_pred_lg = y_test.copy()
y_pred_lg["outtrn_cumsum"] = y_pred_lg_0
y_pred_lg["HeatingEnergyUsage_cumsum"] = y_pred_lg_1



# Calculate RMSE between the predictions and actual 'y' values
def calculate_rmse(targets, predictions):
    """
    Calculate the Root Mean Squared Error (RMSE) between predicted and target values.

    :param predictions: Predicted values.
    :type predictions: array-like
    :param targets: Target values.
    :type targets: array-like
    :return: RMSE value.
    :rtype: float
    """
    from sklearn.metrics import mean_squared_error
    return np.sqrt(mean_squared_error(targets, predictions))


# Calculate r2_score between the predictions and actual 'y' values
def calculate_R2_score(y_test,y_pred):
    from sklearn.metrics import r2_score
    return r2_score(y_test, y_pred)


# rmse_1 = calculate_rmse(y_test, y_pred_1)
# r2score_1 = calculate_R2_score(y_test, y_pred_1)

# rmse = calculate_rmse(y_test, y_pred)
# r2score = calculate_R2_score(y_test, y_pred)

# rmse_rf = calculate_rmse(y_test, y_pred_rf)
# r2score_rf = calculate_R2_score(y_test, y_pred_rf)

rmse_lg = calculate_rmse(y_test, y_pred_lg)
r2score_lg = calculate_R2_score(y_test, y_pred_lg)

# ------------------------------------------------
### OUTPUT ###
# print("RMSE_1:", rmse_1)
# print("R2_score_1:", r2score_1)

# print("RMSE:", rmse)
# print("R2_score:", r2score)

# print("RMSE_rf:", rmse_rf)
# print("R2_score_rf:", r2score_rf)

print("RMSE_lg:", rmse_lg)
print("R2_score_lg:", r2score_lg)
```
